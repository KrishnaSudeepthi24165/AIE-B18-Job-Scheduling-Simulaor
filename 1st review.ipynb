##### Priority Queue

* A queue that arranges elements based on their priority values.
* Each element has a priority value associated with it.
* When we add an item, it is inserted in a position based on its priority value.
* Elements with higher priority values are retrieved before elements with lower priority values.
* If two elements have the same priority, they are served according to their order in the queue.
##### Types of priority queue:
1. Ascending Order Priority Queue
    - The element with least value has the highest priority.
2. Descending order Priority Queue
    - The element with highest value has the highest priority.

##### How to Implement Priority Queue? 

* Arrays
* Linked list
* Binary Heap data structure

##### Why Priority Queue? 

- It helps to access the elements in a faster way without having to search through the entire queue.
- The ordering of elements in a Priority Queue is done dynamically.
- Included in real-time systems where time is of the essence since we retrieve the highest priority element.

##### Which implementation is better?
$  Binary \ Heap \ Data \ Structure$ 
##### Why?


In binary heaps, we have easy access to the min (in min heap) or max (in max heap)
###### But what is a binary heap?
1. It is a Complete Tree.
2. * Min Binary Heap :  Root must be minimum. Same property must be followed by the whole tree.
   * Max Binary Heap :  Root must be maximum. Same property must be followed by the whole tree.
